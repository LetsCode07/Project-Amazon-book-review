```python
#importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urlencode
import csv
```

```python
#5 review pages for a book named as Think straight change your thoughts

list_of_urls = ['https://www.amazon.in/Think-Straight-Change-your-thoughts/product-reviews/0143452134/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews',
'https://www.amazon.in/Think-Straight-Change-your-thoughts/product-reviews/0143452134/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=2',
'https://www.amazon.in/Think-Straight-Change-your-thoughts/product-reviews/0143452134/ref=cm_cr_getr_d_paging_btm_next_3?ie=UTF8&reviewerType=all_reviews&pageNumber=3',
'https://www.amazon.in/Think-Straight-Change-your-thoughts/product-reviews/0143452134/ref=cm_cr_getr_d_paging_btm_next_4?ie=UTF8&reviewerType=all_reviews&pageNumber=4',
'https://www.amazon.in/Think-Straight-Change-your-thoughts/product-reviews/0143452134/ref=cm_cr_getr_d_paging_btm_next_5?ie=UTF8&reviewerType=all_reviews&pageNumber=5']
```

```python
# Retrieve each of the url's HTML data and convert the data into a beautiful soup object.  
# Find, extract and store reviewer names and review text into a list.

names = []
reviews = []
data_string = ""

for url in list_of_urls: 
    params = {'api_key': "39eb2e0779011881f4f109da9e3dbd26", 'url': url}
    response = requests.get('http://api.scraperapi.com/',   params=urlencode(params))
    soup = BeautifulSoup(response.text, 'html.parser')

    for item in soup.find_all("span", class_="a-profile-name"):
      data_string = data_string + item.get_text()
      names.append(data_string)
      data_string = ""  
    
    for item in soup.find_all("span", {"data-hook": "review-body"}):
      data_string = data_string + item.get_text()
      reviews.append(data_string)
      data_string = ""
```

```python
# Create the dictionary.
reviews_dict = {'Reviewer Name': names, 'Reviews': reviews}
# Print the lengths of each list.  
print(len(names), len(reviews))
```

```python
# Create a new dataframe.
df = pd.DataFrame.from_dict(reviews_dict, orient='index')
df.head()
```

```python
# Delete all the columns that have missing values.
df.dropna(axis=1, inplace=True)
df.head()
```

```python
# Transpose the dataframe.
prod_reviews = df.T
print(prod_reviews.head(10))
```

```python
# Remove special characters from review text.
prod_reviews['Reviews'] = prod_reviews['Reviews'].str.replace('\n','')
prod_reviews.head(5)
```

```python
# Convert dataframe to CSV file.
prod_reviews.to_csv('reviews.csv', index=False, header=True)
```
